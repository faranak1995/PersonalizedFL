```python
def calculate_communication_time(model_size_mb, upload_bandwidth_mbps, download_bandwidth_mbps, latency_sec, rounds):
    upload_time = model_size_mb / (upload_bandwidth_mbps / 8)
    download_time = model_size_mb / (download_bandwidth_mbps / 8)
    comm_time_per_round = upload_time + download_time + latency_sec
    total_comm_time = comm_time_per_round * rounds
    return total_comm_time

# Example
model_size_mb = 10
upload_bandwidth_mbps = 10
download_bandwidth_mbps = 20
latency_sec = 0.05
rounds = 100

comm_time = calculate_communication_time(model_size_mb, upload_bandwidth_mbps, download_bandwidth_mbps, latency_sec, rounds)
print(f"Total Communication Time: {comm_time} seconds")
```
